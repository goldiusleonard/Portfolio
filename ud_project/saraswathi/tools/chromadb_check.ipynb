# Remove Date and Datetime column level

```python
import chromadb
from chromadb.config import Settings

chromadb_setting = Settings(
    chroma_client_auth_provider="chromadb.auth.basic_authn.BasicAuthClientProvider",
    chroma_client_auth_credentials="userdata:userdataChromaPassw0rd",
)

client = chromadb.HttpClient(
    host="52.187.106.203",
    port=8000,
    settings=chromadb_setting,
)

# Specify your collection
collection_name = (
    "devtest4.ADA.banking.columnLevel"  # Replace with your collection name
)
collection = client.get_collection(collection_name)

# Query to find vectors with 'sql_data_type' in ['date', 'datetime']
metadata_to_remove = {"sql_data_type": ["date", "datetime"]}

# Retrieve all vectors from the collection
all_vectors = collection.get(
    where={"sql_data_type": {"$in": metadata_to_remove["sql_data_type"]}}
)

# Remove each vector that meets the criteria
for vector in all_vectors["ids"]:
    collection.delete(ids=[vector])

print("Vectors with sql_data_type 'date' or 'datetime' have been removed.")
```

# Update Column Level Metadata

```python
import chromadb
from chromadb.config import Settings

chromadb_setting = Settings(
    chroma_client_auth_provider="chromadb.auth.basic_authn.BasicAuthClientProvider",
    chroma_client_auth_credentials="userdata:userdataChromaPassw0rd",
)

client = chromadb.HttpClient(
    host="52.187.106.203",
    port=8000,
    settings=chromadb_setting,
)

# Create or retrieve a collection
collection = client.get_or_create_collection("liberty.ADA.banking.columnLevel")

ids_edit_list = ["id422-MySQL-ada_banks_data_asset.rfm_projection_new.rfm_month"]


# Function to update metadata
def update_metadata(collection, record_id, new_metadata):
    # Retrieve the existing embedding for the record
    record = collection.get(
        ids=[record_id], include=["embeddings", "metadatas", "documents"]
    )
    embedding = record["embeddings"][0]

    # Delete the old record
    collection.delete(ids=[record_id])

    # Add the embedding back with the new metadata
    collection.add(embeddings=[embedding], ids=[record_id], metadatas=[new_metadata])


# Update metadata for a specific record (e.g., "id1")

result_collection = collection.get(
    ids=ids_edit_list, include=["embeddings", "metadatas", "documents"]
)

for id, embedding, metadata in zip(
    result_collection["ids"],
    result_collection["embeddings"],
    result_collection["metadatas"],
):
    new_metadata = metadata.copy()
    new_metadata["10_samples"] = (
        new_metadata["10_samples"].replace("2018", "2023").replace("2019", "2024")
    )
    new_metadata["all_samples"] = (
        new_metadata["all_samples"].replace("2018", "2023").replace("2019", "2024")
    )
    # new_metadata["avg_value"] = 2023
    new_metadata["max_value"] = (
        new_metadata["max_value"].replace("2018", "2023").replace("2019", "2024")
    )
    new_metadata["min_value"] = (
        new_metadata["min_value"].replace("2018", "2023").replace("2019", "2024")
    )
    new_metadata["universal_all_samples"] = new_metadata["all_samples"].replace(
        "2018", "2023"
    )
    # display(new_metadata)
    update_metadata(collection, id, new_metadata)

# # Verify the metadata update by fetching the updated record
# updated_record = collection.get(ids=["id1"])
# print("Updated metadata:", updated_record['metadatas'])
```
